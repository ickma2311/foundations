## Key Concepts

Let's explore the same linear system from two different geometric perspectives using a concrete example.

### Linear Algebra Introduction
- Linear algebra studies systems of linear equations and matrices.
- The central object: the matrix, representing both equations (rows) and vectors (columns).

### Row Picture
- Each equation is a line (in 2D) or plane (in higher dimensions).
- The solution corresponds to the intersection point(s) of those lines/planes.

**Example**: Consider the system:
$$\begin{align}
x + 2y &= 5 \\
3x + 4y &= 6
\end{align}$$

**Row Picture Interpretation**:
- Equation 1: Line $x + 2y = 5$ (slope = -1/2, y-intercept = 5/2)
- Equation 2: Line $3x + 4y = 6$ (slope = -3/4, y-intercept = 3/2)
- Solution: Intersection point where both equations are satisfied

### Column Picture
- Rewrite the system as a linear combination of columns:
$$A \cdot x = b \quad \Leftrightarrow \quad x_1 \cdot a_1 + x_2 \cdot a_2 = b$$
- A solution exists if and only if $b$ lies in the column space (span of columns).

**Example**: Same system rewritten:
$$x \begin{bmatrix}1 \\ 3\end{bmatrix} + y \begin{bmatrix}2 \\ 4\end{bmatrix} = \begin{bmatrix}5 \\ 6\end{bmatrix}$$

**Column Picture Interpretation**:
- Column 1: Vector $\begin{bmatrix}1 \\ 3\end{bmatrix}$ scaled by $x$
- Column 2: Vector $\begin{bmatrix}2 \\ 4\end{bmatrix}$ scaled by $y$  
- Question: Can we combine these vectors to reach $\begin{bmatrix}5 \\ 6\end{bmatrix}$?
- Solution: Find coefficients $x, y$ such that the linear combination equals $b$

### Possible Outcomes of a Linear System
- **Unique solution**: intersection at one point.
- **No solution**: parallel lines (inconsistent system).
- **Infinitely many solutions**: overlapping lines (dependent equations).

### Matrix Notation
- Compact representation:
$$A \begin{bmatrix}x \\ y\end{bmatrix} = b$$
- Bridges algebraic computation with geometric interpretation.

### Exercise 3: System Types
Give an example of a system of two equations in two unknowns that has:
- (a) a unique solution
- (b) no solution
- (c) infinitely many solutions

```python
# TODO: Create examples for each case
import numpy as np
import matplotlib.pyplot as plt

# (a) Unique solution
A = np.array([[1,2],[3,4]])
b = np.array([5,6])
print("Case (a) - Unique solution:")
print("Solution:", np.linalg.solve(A,b))

# (b) No solution  
A = np.array([[1,2],[2,4]])
b = np.array([5,6])
print("\nCase (b) - No solution:")
try:
    print(np.linalg.solve(A,b))
except np.linalg.LinAlgError as e:
    print("Error:", e)

# (c) Infinitely many solutions
A = np.array([[1,2],[2,4]])
b = np.array([5,10])  # b = 5 * [1,2], so b is in span
print("\nCase (c) - Infinitely many solutions:")
try:
    print(np.linalg.solve(A,b))
except np.linalg.LinAlgError as e:
    print("Error:", e)

# Create subplots for all three cases
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Case (a): Unique solution
ax = axes[0]
A_a = np.array([[1,2],[3,4]])
b_a = np.array([5,6])
solution = np.linalg.solve(A_a, b_a)
x,y = solution[0],solution[1]

# Plot using arrow style like Exercise 2
ax.arrow(0,0,x*A_a[0,0],x*A_a[1,0],head_width=0.1,head_length=0.1,color='blue',label=f'{x:.1f}*a1')
ax.arrow(x*A_a[0,0],x*A_a[1,0],y*A_a[0,1],y*A_a[1,1],head_width=0.1,head_length=0.1,color='green',label=f'{y:.1f}*a2')
ax.arrow(0,0,b_a[0],b_a[1],head_width=0.1,head_length=0.1,color='red',label='b')

ax.grid(True)
ax.set_aspect('equal')
ax.legend(fontsize=8)
ax.set_title('Case (a): Unique Solution\nNon-parallel columns')
ax.set_xlim(-1, 6)
ax.set_ylim(-1, 7)

# Case (b): No solution
ax = axes[1]
A_b = np.array([[1,2],[2,4]])
b_b = np.array([5,6])

ax.quiver(0, 0, A_b[0,0], A_b[1,0], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.005, label='a1 = [1,2]')
ax.quiver(0, 0, A_b[0,1], A_b[1,1], angles='xy', scale_units='xy', scale=1, 
          color='green', width=0.005, label='a2 = [2,4]')
ax.quiver(0, 0, b_b[0], b_b[1], angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.005, label='b = [5,6]')

# Show the span (line through parallel columns)
t = np.linspace(-2, 6, 100)
span_x = t * A_b[0,0]
span_y = t * A_b[1,0]
ax.plot(span_x, span_y, 'k--', alpha=0.5, label='Span of columns')
ax.text(b_b[0]+0.2, b_b[1]+0.2, 'b NOT\non span!', fontsize=10, color='red')

ax.grid(True)
ax.set_aspect('equal')
ax.legend(fontsize=8)
ax.set_title('Case (b): No Solution\nParallel columns, b not in span')
ax.set_xlim(-1, 6)
ax.set_ylim(-1, 7)

# Case (c): Infinitely many solutions
ax = axes[2]
A_c = np.array([[1,2],[2,4]])
b_c = np.array([5,10])  # b = 5 * [1,2]

ax.quiver(0, 0, A_c[0,0], A_c[1,0], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.005, label='a1 = [1,2]')
ax.quiver(0, 0, A_c[0,1], A_c[1,1], angles='xy', scale_units='xy', scale=1, 
          color='green', width=0.005, label='a2 = [2,4]')
ax.quiver(0, 0, b_c[0], b_c[1], angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.005, label='b = [5,10]')

# Show the span and multiple solutions
t = np.linspace(-2, 6, 100)
span_x = t * A_c[0,0]
span_y = t * A_c[1,0]
ax.plot(span_x, span_y, 'k--', alpha=0.5, label='Span of columns')

# Show different linear combinations that reach b
# Solution 1: 5*a1 + 0*a2 = b
ax.quiver(0, 0, 5*A_c[0,0], 5*A_c[1,0], angles='xy', scale_units='xy', scale=1,
          color='purple', alpha=0.7, label='5*a1 + 0*a2')
# Solution 2: 3*a1 + 1*a2 = b  
ax.quiver(0, 0, 3*A_c[0,0], 3*A_c[1,0], angles='xy', scale_units='xy', scale=1,
          color='orange', alpha=0.7, label='3*a1 + 1*a2')
ax.quiver(3*A_c[0,0], 3*A_c[1,0], 1*A_c[0,1], 1*A_c[1,1], angles='xy', scale_units='xy', scale=1,
          color='orange', alpha=0.7)

ax.text(b_c[0]+0.2, b_c[1]+0.2, 'b ON span!\nInfinite solutions', fontsize=10, color='red')

ax.grid(True)
ax.set_aspect('equal')
ax.legend(fontsize=8)
ax.set_title('Case (c): Infinite Solutions\nParallel columns, b in span')
ax.set_xlim(-1, 6)
ax.set_ylim(-1, 11)

plt.tight_layout()
plt.show()

print("\nSummary:")
print("Case (a): Non-parallel columns  unique solution")
print("Case (b): Parallel columns, b not in span  no solution") 
print("Case (c): Parallel columns, b in span  infinite solutions")
```
